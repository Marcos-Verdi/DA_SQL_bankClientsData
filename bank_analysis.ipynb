```python
credit_limit_range = pd.read_sql_query("""
SELECT
MIN(Credit_Limit) AS smalles_credit_limit, MAX(Credit_Limit) AS largest_credit_limit
FROM bankchurners
""", conn)

credit_limit_range.style.hide_index()
```

```python
count_by_credit_category = pd.read_sql_query("""
SELECT
CASE
WHEN Credit_Limit BETWEEN 0 AND 5000 THEN '0-5000'
WHEN Credit_Limit BETWEEN 5001 AND 10000 THEN '5000-10000'
WHEN Credit_Limit BETWEEN 10001 AND 15000 THEN '10000-15000'
WHEN Credit_Limit BETWEEN 15001 AND 20000 THEN '15000-20000'
WHEN Credit_Limit BETWEEN 20001 AND 25000 THEN '20000-25000'
WHEN Credit_Limit BETWEEN 25001 AND 30000 THEN '25000-30000'
ELSE '30000 +' END AS credit_category,
COUNT(*) AS count
FROM bankchurners
GROUP BY credit_category
ORDER BY count DESC
""", conn)

count_by_credit_category.style.hide_index()
```

```python
count_transactions_range = pd.read_sql_query("""
SELECT
MIN(Total_Trans_Ct) AS least_count_transaction, MAX(Total_Trans_Ct) AS greatest_count_transaction
FROM bankchurners
""", conn)

count_transactions_range.style.hide_index()
```

```python
count_by_transaction_category = pd.read_sql_query("""
SELECT
CASE
WHEN Total_Trans_Ct BETWEEN 0 AND 25 THEN '0-25'
WHEN Total_Trans_Ct BETWEEN 26 AND 50 THEN '25-50'
WHEN Total_Trans_Ct BETWEEN 51 AND 100 THEN '50-100'
ELSE '100 +' END AS transaction_ct_category,
COUNT(*) AS count
FROM bankchurners
GROUP BY transaction_ct_category
ORDER BY count DESC
""", conn)

count_by_transaction_category.style.hide_index()
```

## Does age play a role in relationship count? revolving balance? Total transaction count and amount? average utilization ratio?

## Do months on book lead to higher transaction amounts? amount change? and higher utilization ratio?

```python
months_on_book_range = pd.read_sql_query("""
SELECT
MIN(Months_on_book) AS min_months_on_book, MAX(Months_on_book) AS max_months_on_book
FROM bankchurners
""", conn)

months_on_book_range.style.hide_index()
```

```python
transaction_amount_per_mobook = pd.read_sql_query("""
SELECT
CASE
WHEN Months_on_book BETWEEN 12 AND 18 THEN '12-18'
WHEN Months_on_book BETWEEN 19 AND 24 THEN '18-24'
WHEN Months_on_book BETWEEN 25 AND 30 THEN '24-30'
WHEN Months_on_book BETWEEN 31 AND 36 THEN '30-36'
WHEN Months_on_book BETWEEN 37 AND 42 THEN '36-42'
WHEN Months_on_book BETWEEN 43 AND 48 THEN '42-48'
WHEN Months_on_book BETWEEN 49 AND 54 THEN '48-54'
ELSE '54<' END AS months_on_book_category,
AVG(Total_Trans_Amt) AS avg_transaction_amount
FROM bankchurners
GROUP BY months_on_book_category
ORDER BY months_on_book_category
""", conn)

transaction_amount_per_mobook.style.hide_index()
```

```python
utilization_ratio_per_mobook = pd.read_sql_query("""
SELECT
CASE
WHEN Months_on_book BETWEEN 12 AND 18 THEN '12-18'
WHEN Months_on_book BETWEEN 19 AND 24 THEN '18-24'
WHEN Months_on_book BETWEEN 25 AND 30 THEN '24-30'
WHEN Months_on_book BETWEEN 31 AND 36 THEN '30-36'
WHEN Months_on_book BETWEEN 37 AND 42 THEN '36-42'
WHEN Months_on_book BETWEEN 43 AND 48 THEN '42-48'
WHEN Months_on_book BETWEEN 49 AND 54 THEN '48-54'
ELSE '54<' END AS months_on_book_category,
AVG(Avg_Utilization_Ratio)*100 AS avg_utilization_ratio
FROM bankchurners
GROUP BY months_on_book_category
ORDER BY months_on_book_category
""", conn)

utilization_ratio_per_mobook.style.hide_index()
```
